```python
os.chdir('C:\\Users\\malowa\\Desktop\\R work\\matplotlib_tutorial-master')
with open('fifa_data.csv',errors='ignore') as csv_file:  #it worked from stackoverflow
    csv_reader = csv.DictReader(csv_file)
    height_counter=Counter()
    for row in csv_reader:
        height_counter.update(row['Height'].split(';'))

height=[]
population=[]

for item in height_counter.most_common(14):
    height.append(item[0])
    population.append(item[1])
    
print(height)
print(population)

```

```python
plt.bar(height,population,label='Heights')
plt.xlabel("Height in ft'inches",fontdict={'fontsize':15})
plt.ylabel("Number of FIFA Players",fontdict={'fontsize':15})
plt.title("Number of FIFA players in 2018 Against Their Heights",fontdict={'fontname':'Times New Roman','fontsize':23,'fontweight':'bold'})
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[8,8])
height.reverse()
population.reverse()

plt.barh(height,population,label='Heights')

plt.xlabel("Height in ft'inches",fontdict={'fontsize':15})
plt.ylabel("Number of FIFA Players",fontdict={'fontsize':15})
plt.title("Number of FIFA players in 2018 Against Their Heights",fontdict={'fontname':'Times New Roman','fontsize':23,'fontweight':'bold'})
plt.legend()
plt.savefig('Heights of FIFA players.png')
plt.show()
```

```python
import csv
from collections import Counter
import numpy as np
```

```python
with open('fifa_data.csv',errors='ignore') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    jersey_counter=Counter()
    
    for row in csv_reader:
        jersey_counter.update(row['Jersey Number'].split(';'))

jersey_number=[]
popularity=[]

for item in jersey_counter.most_common(30):
    jersey_number.append(item[0])
    popularity.append(item[1])
    

print(jersey_number)
print(popularity)
   
    
```

```python
plt.figure(figsize=[12,8])
jersey_number.reverse()
popularity.reverse()
plt.barh(jersey_number,popularity,color='#5ce1ff')
plt.ylabel('Jersey Numbers',fontdict={'fontsize':15})
plt.xlabel('Popularity',fontdict={'fontsize':15})
plt.title('The Popularity of Jersey Numbers in FIFA 2018',fontdict={'fontname':'Times New Roman','fontsize':23,'fontweight':'bold'})
plt.savefig('The Popularity of Jersey Numbers in FIFA 2018.png')
plt.grid(True)
plt.show()
```

```python
fif=pd.read_csv('fifa_data.csv')

fif=fif.fillna(0)

fif['Jersey Number']=fif['Jersey Number'].astype(int)

fif['Jersey Number']=fif['Jersey Number'].astype(str)

jerseys=fif['Jersey Number']

counter=Counter(jerseys).most_common(30)

jerseyno=[]

liked=[]

for i in counter:
    jerseyno.append(i[0])
    liked.append(i[1])
    
```
